```python
import os
import pandas as pd
from PIL import Image
import time
import torch
import torchvision
from torch.utils.data import Dataset
from torchvision import transforms
import albumentations
import albumentations.pytorch
from matplotlib import pyplot as plt
import cv2
import numpy as np
```

```python
class ImageQualityDataset(Dataset):
    def __init__(self, filepaths, labels, transform=None):
        self.file_paths = filepaths
        self.labels = labels
        self.transform = transform
        
    def __len__(self):
        return len(self.file_paths)
    
    def __getitem__(self, idx):
        filepath = self.file_paths[idx]
        label = self.labels[idx]
        
        image = cv2.imread(filepath)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        if self.transform:
            augmented_image = self.transform(image=image)
            image = augmented_image['image']
        return image, label
```

```python
albumnation_transform = albumentations.Compose([
                            albumentations.RandomRotate90(),
                            albumentations.ShiftScaleRotate(shift_limit=0.0625, scale_limit=0.2, rotate_limit=5, p=0.5),
                            albumentations.pytorch.transforms.ToTensorV2()
                            ])
```

```python
image_root_dir = r"D:\Gramener\NHA_PoC\image_quality_check_19_07\train"
# doc_class = 'good'
for doc_class in ['good', 'bad']:
    file_paths = [os.path.join(image_root_dir, doc_class, i) for i in os.listdir(os.path.join(image_root_dir, doc_class))]
    albumentations_dataset = ImageQualityDataset(
        filepaths=file_paths,
        labels=[doc_class]*len(file_paths),
        transform=albumnation_transform,
    )
    for i in range(5):
        for image_no in range(len(albumentations_dataset)):
            sample_image, sample_label = albumentations_dataset[image_no]
            converted_image = transforms.ToPILImage()(sample_image)
            final_image_path = os.path.join(image_root_dir, doc_class+'_aug', doc_class+'_'+str(image_no)+'_'+str(i)+'.jpg')
            converted_image.save(final_image_path)
```
